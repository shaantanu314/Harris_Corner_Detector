# Harris corner Detector

```python
from scipy import signal
from scipy import misc
from PIL import Image
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as pplt
import matplotlib.image as mpimg
```

Lets import the image and then convert it to Grayscale since we are just interested in the intensity of the image to calculate the gradients.



```python
input_img = mpimg.imread('image.jpg')
input_img = np.array(input_img)
pplt.imshow(input_img)
pplt.show()

```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

#  grayscale image array
img = rgb2gray(input_img) 
# img = np.dot(input_img[...,:3], rgb_weights)
```

```python
pplt.imshow(img, cmap=pplt.get_cmap("gray"))
pplt.show()
```

Calculating the gradients of the image


```python
# print(img.shape)
dy , dx = np.gradient(img)
fig = pplt.figure(figsize=(10, 7))
fig.add_subplot(1, 2, 1)
pplt.imshow(dy)
pplt.title("Gradient along y direction")
fig.add_subplot(1, 2, 2)
pplt.imshow(dx)
pplt.title("Gradient along x direction")

```

```python
# Parameters for the algorithm
window_size = 5
k = 0.04
threshold = 100000000

offset = int(window_size/2)

height = img.shape[0]
width = img.shape[1]
```

```python
Ixx = dx**2
Iyy = dy**2
Ixy = dy*dx
corner_list = []
edge_list = []
output_img_corner =  np.empty_like(input_img)
output_img_corner[:] = input_img
output_img_edge = np.empty_like(input_img)
output_img_edge[:] = input_img
for y in range(offset, height-offset):
        for x in range(offset, width-offset):
            
            #Values of sliding window
            start_y = y - offset
            end_y = y + offset + 1
            start_x = x - offset
            end_x = x + offset + 1
            
            #The variable names are representative to 
            #the variable of the Harris corner equation
            windowIxx = Ixx[start_y : end_y, start_x : end_x]
            windowIxy = Ixy[start_y : end_y, start_x : end_x]
            windowIyy = Iyy[start_y : end_y, start_x : end_x]
            
            #Sum of squares of intensities of partial derevatives 
            Sxx = windowIxx.sum()
            Sxy = windowIxy.sum()
            Syy = windowIyy.sum()

            #Calculate determinant and trace of the matrix
            det = (Sxx * Syy) - (Sxy**2)
            trace = Sxx + Syy
            
            #Calculate r for Harris Corner equation
            r = det - k*(trace**2)

            if r > threshold:
                corner_list.append([x, y, r])
            if r< -1000000:
                edge_list.append([x,y,r])
#                 output_img[y,x] = (0,0,255)
```

```python
# print(corner_list)
for edge in edge_list:
    output_img_edge[edge[1],edge[0]] = (0,255,0)
    
for corner in corner_list:
    output_img_corner[corner[1],corner[0]] = (0,0,255)
    
```

```python
fig = pplt.figure(figsize=(10, 7))
fig.add_subplot(1, 2, 1)
pplt.imshow(output_img_edge)
pplt.title("Edge Detection")
fig.add_subplot(1, 2, 2)
pplt.imshow(output_img_corner)
pplt.title("Corner Detection")
```

So we can see the green edges and the blue corners have been detected using the Harris corner detector. Here we have used a uniform window , but we can also use a gaussian window.
